# EXPLANATORY DATA VISUALIZATION 

After doing some data exploration on the titanic dataset, The intended finding of how variables like passenger class, gender of the passenger, age of the passenger could affect result of survivor.

It is well observed that the passenger class had correlation with the outcome of survivor. Another observation was many males succumbed to the tragedy comapared to females.

Another interesting feature was how passenger's age related with the Pclass they aboard. 

```python
#loading of libraries 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv("titanic_data.csv")
titanic.head()
```

# 2.0 Explanatory visualization 1

passengers who survived more according to the class they boarded were FIrst class, followed by Second and third class had few number of survivors


```python
#we can make an observation that the status of the class determined the number of survivors
sns.barplot(x = 'Pclass', y='Survived',data=titanic)
plt.title("Status of class against number of Survivors");
```

# 2.1 EXPLANATORY VISUALIZATION 2

For each of the passenger class an area is plotted that shows an approximate age for passenger class. Those passengers who were on Third class majority of them were below 30 years.

```python
base_color = sns.color_palette()[0]
sns.violinplot(data=titanic, x='Pclass', y='Age', color = base_color, inner = None);
plt.xticks(rotation = 15)
plt.title('Pclass against age distribution');
```

# 2.2 EXPLANATORY VISUALIZATION 3
The number of male who survived was 109 passengers, while the females who survived were 231.

More females survived than males.

```python
sns.countplot(data = titanic, x = 'Sex', hue = 'Survived');
plt.title('Distribution of Gender survivor');
```

# 2.3 EXPLANATORY VISUALIZATION 4
The qualitative and quantitative relationship between the four variables can be seen. The number of male who were aged 20-40 did not survive. From the obsevation blue is many than orange, showing males died more than females.

```python
f, ax = plt.subplots(1,1, figsize=(10, 5))
sns.scatterplot(x="Fare", y="Age", hue="Survived", size="Sex", data=titanic, ax=ax);
plt.title('Relationship of Age, Fare, Survived and gender');
```
